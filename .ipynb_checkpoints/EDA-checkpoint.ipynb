# 3) Insight analysis based on outcome

```python
# Investigate age against outcome

# Group ages for more volume
# Define the age ranges
age_bins = [18, 25, 35, 45,55,65,float('inf')]

#Labels for the age groups
age_labels = ['18-25', '26-35', '36-45', '46-55','56-65','66+']

# Create a new column 'age_group' with the age groups
fullDataset['age_group'] = pd.cut(fullDataset['age'], bins=age_bins, labels=age_labels, right=False)

# Take cross tabulation and see volumes
props = pd.crosstab(fullDataset['age_group'],fullDataset['y'], normalize='index')['yes']
print(pd.crosstab(fullDataset['age_group'],fullDataset['y']))

# Create a bar chart from the normalized cross-tabulation
props.plot(kind='bar', stacked=True)

# Add labels and a legend
plt.xlabel('Age group')
plt.ylabel('Proportion')
# Show the plot
plt.show()

# Seems like younger and older population more likely.
# Younger makes sense as generally want to save/invest when young and older usually have more money lying around when retired so put away for loved ones.
```

```python
# Take cross tabulation and see volumes
props = pd.crosstab(fullDataset['job'],fullDataset['y'], normalize='index')['yes'].sort_values()
print(pd.crosstab(fullDataset['job'],fullDataset['y']))

# Create a bar chart from the normalized cross-tabulation
props.plot(kind='bar', stacked=True)

# Add labels and a legend
plt.xlabel('Job type')
plt.ylabel('Proportion')
# Show the plot
plt.show()

# Students have the highest success with fairly low volumes
# Next highest is retired.
# Top 3 categories of people all do not work, maybe have more time to answer the phone?
# Blue collar are lowest likely.
```

```python
# Convert outcomes to binary 1/0 
fullDataset['outcome'] = pd.Series(np.where(fullDataset.y.values == 'yes', 1, 0))

months = ["jan", "feb", "mar", "apr", "may", "jun", 
          "jul", "aug", "sep", "oct", "nov", "dec"]
fullDataset["month"] = pd.Categorical(fullDataset['month'], categories=months, ordered=True)

# Plot number of calls made across months
fullDataset.groupby('month').size().plot(kind='bar')
plt.show()

# Plot conversion rate across months
fullDataset.groupby('month')['outcome'].mean().plot(kind='bar')
plt.show()

# No obvious seasonality/pattern showing. Mar, Sep, Oct and Dec have high success rates but volume of calls made are very small.
```

# Assumptions
1) In line with the task, the analysis will be run on more recent data. Ideally I would have liked to include the year information and analysed data over time however this would require manual input of year information which would go against the task criterion.
2) Cellular and telephone are the same. The differences are ambiguous therefore they have been treated as the same until further clarification

# 4) Learnings/profiling
1) Younger (<25) and older (55+) population generally have a higher conversion rate. Younger population volumes are very low so inconclusive however older population have higher volumes. Ages 36-55 have a lower than average conversion rate of about 9-10% with volumes to make this observation conclusive. Recommend target ages other than 36-55 y/o.
2) Blue-collar jobs have a really low conversion rate with large volumes to make observation conclusive. Blue-collar conversation rate is ~7% which is a lot lower than average even though majority calls made to blue-collar customers. Recommed exclude blue-collar customers.
3) Grouping marital status into 2 categories, married vs unmarried (single and divorced), unmarried have a slightly higher conversion rate of 12-14% whereas married have conversion rate of ~10%. Not enough to exclude married. Recommend target higher proportion of unmarried.
4) Tertiary education has a much higher conversion rate of ~15% compared to primary at 9%. Other categories float around the average. Recommend exclude primary education candidates.
5) Customers without defaults have an almost double conversion rate. Good customer outcome to exclude defaults as this money may be needed to pay off CCJ's. Recommend exclude defaults.
6) As average balance increases, conversion rate increases however volumes also decrease. Customers with balance >100 achieve conversion rate of 13% whilst balance <100 has conversion rate of 8%. 60% increase in conversion for customer with balance >100. Recommend excluding customers with avg balance < 100.
7) Customers with a housing loan or other loan have a much lower conversion rate between 7-8% compared to customers without at 13-16%. Recommend excluding customers with a loan.

# 5) Validation

```python
validationPopulation = fullDataset[(fullDataset['job'] != 'blue-collar') & (fullDataset['education'] != 'primary') & (fullDataset['default'] == 'no') & (fullDataset['balance'] > 100) & (fullDataset['housing'] == 'no') & (fullDataset['loan'] == 'no') ]
print(validationPopulation['outcome'].mean())
print(len(validationPopulation))
```

# Summary
Given the above mentioned exclusions, the population volume still holds quite high at 10186 however the conversion rate increases to 22% on the training set. This will have to be tested on a validation population and ensure the figures still hold

## Pro's of the above analysis
1) We full understand the exclusions and can put some sort of reasoning/rationale behind them. An example would be those with a lower balance may not be able to afford the fixed-term deposit product so the recommendation holds some reasoning.
2) We better understand customer profiles and can target based on these profiles.

## Cons of the above analysis
1) We are unable to see how behaviour is changing over time. It would be good to see how changes are made YoY therefore including full dates would be really helpful.
2) The analysis are made on training data, we would want to validate these results by having a control and experiment population.
3) The data is quite biased in some characteristics, it would be good to have more calls to those characteristics where data is sparse to improve our understandings.
